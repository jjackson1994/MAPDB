```python
# loop for multiple faces
import time
from pathlib import Path, PurePath

boxsize = []
def preProcess(n_images = 10,save_to = 'data/faces/'):
    def scaleImage(image, max_dim = 500):
            shape = np.shape(image)
            max_side = np.argmax(shape)
            scale_factor = max_dim / shape[max_side]
            image_scaled = transform.rescale(image,scale_factor, multichannel = True, mode = 'constant', cval = 0)

            xpad1,xpad2 = (0,0)
            ypad1,ypad2 = (0,0)
            if max_side == 0:
                ypad = (500 - np.shape(image_scaled)[1])
                ypad1 = int(np.floor(ypad/2 + ypad%2))
                ypad2 = int(np.floor(ypad/2))

            if max_side == 1:
                xpad = (500 - np.shape(image_scaled)[0])
                xpad1 = int(np.floor(xpad/2 + xpad%2))
                xpad2 = int(np.floor(xpad/2))

            image_padded = np.pad(image_scaled, pad_width = [(xpad1,xpad2),(ypad1,ypad2),(0,0)], mode = 'constant', constant_values = (0,0))
            return image_padded
    
    def extractFace(im,xp,yp,wp,hp,face_no):
            im_width = np.shape(im)[1]
            im_height = np.shape(im)[0]
            # Center of Box
            xc = im_width * xp
            yc = im_height * yp
            # Width and height of box
            w = im_width * wp
            h = im_height * hp
            boxsize.append(w*h)
            # Rows ans Columns of Box
            rows = (int(yc-h/2),int(yc+h/2))
            columns = (int(xc-w/2),int(xc+w/2))
            # crop to box
            im_face = im[rows[0]:rows[1],columns[0]:columns[1],:]
            im_scaled = scaleImage(im_face)
            #io.imshow(im_scaled)
            #io.show()
            #io.imsave(fname = Path(save_to+'face_'+face_no), arr = im_scaled)
    face_no = 1;             
    for i in range(1,n_images+1):
        if (i%100==0): print(i)
        try: im = io.imread(Path('data/images/Image_'+str(i)+'.jpg'))    
        except:
            try: im = io.imread(Path('data/images/Image_'+str(i)+'.jpeg'))
            except:
                try: im = io.imread(Path('data/images/Image_'+str(i)+'.png'))
                except:
                    try: io.imread(Path('data/images/Image_'+str(i)+'.JPG'))
                    except:
                        print('File Not Found')
                        continue
        #Some images have a fourth channel filled with ones, this can break the padding
        im = im[:,:,0:3]
        
#         try: 
#             _,xp,yp,wp,hp = np.loadtxt(Path('data/labels/Image_'+str(i)+'.txt'))
#             extractFace(im,xp,yp,wp,hp, face_no = str(face_no))
#             face_no = face_no + 1
#         except:
#             label = np.loadtxt(Path('data/labels/Image_'+str(i)+'.txt'))
#             for lbl in label:
#                 _,xp,yp,wp,hp = lbl
#                 extractFace(im,xp,yp,wp,hp, face_no = str(face_no))
#                 face_no += 1
                
        label = np.loadtxt(Path('data/labels/Image_'+str(i)+'.txt'))
#         print(type(label[0]))
#         print(label[0])
        if (str(type(label[0])) == "<class 'numpy.float64'>"):
            _,xp,yp,wp,hp = np.loadtxt(Path('data/labels/Image_'+str(i)+'.txt'))
            extractFace(im,xp,yp,wp,hp, face_no = str(face_no))
            face_no = face_no + 1
        elif  (str(type(label[0])) == "<class 'numpy.ndarray'>"):
            for lbl in label:
                _,xp,yp,wp,hp = lbl
                extractFace(im,xp,yp,wp,hp,face_no = str(face_no))
                face_no += 1
        
        
%time
preProcess(2000)
```

```python
# remove faces from photos
# randomly select a part of the photo thats 500x500
# loop for multiple faces
import time
from pathlib import Path, PurePath

randGen = np.random.default_rng(12345)

def getNoface(n_images = 10,save_to = 'data/faces/'):
    def scaleImage(image, max_dim = 500):
            shape = np.shape(image)
            max_side = np.argmax(shape)
            scale_factor = max_dim / shape[max_side]
            image_scaled = transform.rescale(image,scale_factor, multichannel = True, mode = 'constant', cval = 0)

            xpad1,xpad2 = (0,0)
            ypad1,ypad2 = (0,0)
            if max_side == 0:
                ypad = (500 - np.shape(image_scaled)[1])
                ypad1 = int(np.floor(ypad/2 + ypad%2))
                ypad2 = int(np.floor(ypad/2))

            if max_side == 1:
                xpad = (500 - np.shape(image_scaled)[0])
                xpad1 = int(np.floor(xpad/2 + xpad%2))
                xpad2 = int(np.floor(xpad/2))

            image_padded = np.pad(image_scaled, pad_width = [(xpad1,xpad2),(ypad1,ypad2),(0,0)], mode = 'constant', constant_values = (0,0))
            return image_padded
    
    def extractNoface(im,xp,yp,wp,hp):
            im_width = np.shape(im)[1]
            im_height = np.shape(im)[0]
            # Center of Box
            xc = im_width * xp
            yc = im_height * yp
            # Width and height of box
            w = im_width * wp
            h = im_height * hp
            # Rows ans Columns of Box
            rows = (int(yc-h/2),int(yc+h/2))
            columns = (int(xc-w/2),int(xc+w/2))
            # remove face
            im[rows[0]:rows[1],columns[0]:columns[1],:] = 0
            #no face x
            im_noface_size = int(np.min((im_width,im_height))/3)
            nfx = randGen.integers(im_noface_size,im_width - im_noface_size)
            nfy = randGen.integers(im_noface_size,im_height - im_noface_size)
            im_noface = im[nfy-im_noface_size:nfy+im_noface_size, nfx-im_noface_size:nfx+im_noface_size, :]
            im_noface = scaleImage(im_noface)
            return im_noface
    face_no = 1;             
    for i in range(1,n_images+1):
        if (i%100==0): print(i)
        try: im = io.imread(Path('data/images/Image_'+str(i)+'.jpg'))    
        except:
            try: im = io.imread(Path('data/images/Image_'+str(i)+'.jpeg'))
            except:
                try: im = io.imread(Path('data/images/Image_'+str(i)+'.png'))
                except:
                    try: io.imread(Path('data/images/Image_'+str(i)+'.JPG'))
                    except:
                        print('File Not Found')
                        continue
        #Some images have a fourth channel filled with ones, this can break the padding
        im = im[:,:,0:3]
        
#         try: 
#             _,xp,yp,wp,hp = np.loadtxt(Path('data/labels/Image_'+str(i)+'.txt'))
#             extractFace(im,xp,yp,wp,hp, face_no = str(face_no))
#             face_no = face_no + 1
#         except:
#             label = np.loadtxt(Path('data/labels/Image_'+str(i)+'.txt'))
#             for lbl in label:
#                 _,xp,yp,wp,hp = lbl
#                 extractFace(im,xp,yp,wp,hp, face_no = str(face_no))
#                 face_no += 1
                
        label = np.loadtxt(Path('data/labels/Image_'+str(i)+'.txt'))
#         print(type(label[0]))
#         print(label[0])
        if (str(type(label[0])) == "<class 'numpy.float64'>"):
            _,xp,yp,wp,hp = np.loadtxt(Path('data/labels/Image_'+str(i)+'.txt'))
            too_much_black = True
            iterations = 0
            im_noface = extractNoface(im,xp,yp,wp,hp)
            while(too_much_black):
                iterations += 1
                im_noface = extractNoface(im,xp,yp,wp,hp)
                if (np.where(im_noface == 0)[0].size < (np.shape(im_noface)[0]**2)/4): too_much_black = False
                if iterations > 5: 
#                     print('broken')
                    break
#             io.imshow(im_noface)
#             io.show()
            face_no = face_no + 1
            io.imsave(fname = Path('data/noFaces/noFace_'+str(face_no)), arr = im_noface)
        elif  (str(type(label[0])) == "<class 'numpy.ndarray'>"):
            for lbl in label:
                _,xp,yp,wp,hp = lbl
                too_much_black = True
                iterations = 0
                im_noface = extractNoface(im,xp,yp,wp,hp)
                while(too_much_black):
                    iterations += 1
                    im_noface = extractNoface(im,xp,yp,wp,hp)
                    if (np.where(im_noface == 0)[0].size < (np.shape(im_noface)[0]**2)/4): too_much_black = False
                    if iterations > 5: 
#                         print('broken')
                        break
#                 io.imshow(im_noface)
#                 io.show()
                face_no = face_no + 1
                io.imsave(fname = Path('data/noFaces/noFace_'+str(face_no)), arr = im_noface)
        
        
%time
getNoface(2000)
```
